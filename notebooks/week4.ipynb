- Week 4: Why Ito fails - fBM and Young Integration, Study fractional Brownian motion as a natural extension of Brownian motion, focusing on its covariance structure, the role of the Hurst parameter, and the loss of the semimartingale property. Explore why classical Itô integration fails in this setting and how Young integration extends deterministic integration to sufficiently regular paths, illustrating both its power and its limitations.
- Week 5: Rough paths: Introduce the central ideas of rough path theory as a framework for defining differential equations driven by highly irregular signals. Learn how enhancing a path with iterated integrals (such as Lévy areas) restores well-posedness and continuity, providing a pathwise extension of stochastic calculus beyond the semimartingale setting.
